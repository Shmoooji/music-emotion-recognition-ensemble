# **Spectrogram Display Notebook**

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa

def plot_spectrograms(spectrogram_file):
    all_spectrograms = np.load(spectrogram_file)

    for i, spectrogram in enumerate(all_spectrograms):
        plt.figure(figsize=(10, 4))
        librosa.display.specshow(spectrogram, x_axis='time', y_axis='mel', hop_length=250)
        plt.colorbar(format='%+2.0f dB')
        plt.title(f'Log Mel Spectrogram - Segment {i+1}')
        plt.tight_layout()
        plt.show()

spectrogram_dir = "../data/spectrograms/"
file_name = "2_spectrograms.npy"

plot_spectrograms(spectrogram_dir + file_name)
```
